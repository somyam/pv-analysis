```python
from venv import logger
import numpy as np
def convert_age_to_years(age_value, age_code):
    """
    Convert age to years based on age_code
    Handles all FAERS age codes: YR, MON, WK, DY, HR, DEC
    """
    if pd.isna(age_value) or pd.isna(age_code):
        return np.nan
    
    try:
        age_value = float(age_value)
    except (ValueError, TypeError):
        return np.nan
    
    # Convert to years based on age_code
    if age_code == 'YR':
        return age_value  # Already in years
    elif age_code == 'MON':
        return age_value / 12  # Convert months to years
    elif age_code == 'WK':
        return age_value / 52.1775  # Convert weeks to years (365.25/7)
    elif age_code == 'DY':
        return age_value / 365.25  # Convert days to years
    elif age_code == 'HR':
        return age_value / (365.25 * 24)  # Convert hours to years
    elif age_code == 'DEC':
        return age_value * 10  # Decades to years
    else:
        # Unknown age code, assume years as fallback
        logger.warning(f"Unknown age_code '{age_code}', assuming years for age {age_value}")
        return age_value

demo['age_clean'] = demo.apply(
        lambda row: convert_age_to_years(row['age'], row['age_cod']), 
        axis=1
    )
```

```python
def categorize_age_research_standard(age):
    """Age groups from pharmacovigilance research"""
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Pediatric'
    elif age <= 44:
        return 'Young Adult'  # 18-44 from research
    elif age <= 64:
        return 'Middle Aged'  # 45-64 from research
    elif age <= 84:
        return 'Elderly'      # 65-84 from research
    else:
        return 'Very Elderly' # 85+ from research
```

```python
demo['age_group'] = demo['age_clean'].apply(categorize_age_research_standard)
demo.loc[:, ("age_group", "age_clean")]
```

```python
# Sex distribution with renamed labels and actual counts
sex_counts = demo['sex'].value_counts()
sex_counts.index = sex_counts.index.map({'F': 'Female', 'M': 'Male'})

ax = sex_counts.plot(kind='bar', figsize=(10, 6), title='Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(rotation=0)

# Format y-axis to show actual numbers instead of scientific notation
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))

plt.tight_layout()
plt.show()
```

```python
# Age group distribution bar graph
demo['age_group'].value_counts().sort_index().plot(kind='bar', figsize=(10, 6), title='Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
